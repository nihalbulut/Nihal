```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
model=KMeans(n_clusters=6)
```

```python
model=model.fit(df)
```

```python
tahmin=model.predict(df)
```

```python
silhouette_score(df,tahmin)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df.Alcohol, df.Proline, c=tahmin)
plt.xlabel("alcohol")
plt.ylabel("Proline")
plt.colorbar()
plt.grid()
```

```python
allscore=[]
allclusters=[]
uzaklıklar=[]
def clust(kaçtane):
    for i in range(1,kaçtane+1):
        i=i+1
        model=KMeans(n_clusters=i)
        moddel=model.fit(df)
        tahmin=model.predict(df)
        score=silhouette_score(df,tahmin)
        uzaklık=model.inertia_
        print("cluster",i,"score",score)
        allscore.append(score)
        allclusters.append(i)
        uzaklıklar.append(uzaklık)
clust(10)
```

```python
plt.xlabel("cluster")
plt.ylabel("uzaklıklar")
plt.plot(allclusters,uzaklıklar)
```
