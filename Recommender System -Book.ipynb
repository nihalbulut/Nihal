```python
df=pd.merge(df,df1,on="ISBN",how="outer")
```

```python
df.head()
```

```python
df.drop(["Image-URL-S","Image-URL-M","Image-URL-L"],axis=1)
```

```python
df.isnull().sum()
```

```python
import numpy as np
```

```python
grouped=df.groupby("Book-Title").agg({"Book-Rating":[np.size,np.sum,np.mean]})
```

```python
grouped.head()
```

```python
populer=grouped.sort_values(("Book-Rating","mean"),ascending=False)
```

```python
populer.head()
```

```python
toplam=grouped["Book-Rating"]["sum"].sum()
```

```python
populer["percentage"]=populer["Book-Rating","sum"].div(toplam)*100
```

```python
populer.sort_values(("percentage"),ascending=False)
```
